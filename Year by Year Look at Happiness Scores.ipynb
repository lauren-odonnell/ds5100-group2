# The code below shows a scatterplot comparing each of the 6 components with the happiness score (region being the color) to help show where each region ranks and what they are strong/weak in

```python
data["Years"] = data["Year"]
data2 = [data["Year"], data["Years"], data["Economy (GDP per Capita)"],data["Family"], data["Freedom"], data["Health (Life Expectancy)"], data["Trust (Government Corruption)"], data["Generosity"]]
headers = ["Year","Years", "Economy (GDP per Capita)", "Family","Freedom", "Health (Life Expectancy)", "Trust (Government Corruption)", "Generosity" ]
data3 = pd.concat(data2, axis=1, keys=headers)
data4 = data3.groupby(['Year']).mean()
data4 = data4.astype({"Years": int})
```

```python
df2 = pd.DataFrame([[2,3,4]], columns=['A','B','C'])
```

```python
data5 = [data["Year"], data["Years"],data["Happiness Score"]]
headers2 = ["Year", "Years", "Happiness Score"]
data6=pd.concat(data5, axis=1, keys=headers2)
data7= data6.groupby(['Year']).mean()
data7 = data7.astype({"Years": int})
data7
```

```python
data4.plot(x='Years',
        kind='bar',
        figsize=(20,5),
        stacked=False,
        title='Grouped Bar Graph of Scores by Year')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
#hard to interpret, but shows scores by each year for each score
```

```python
low = min(data7['Happiness Score'])-.01
high = max(data7['Happiness Score'])+.01
plt.ylim(low, high)
plt.bar(data7["Years"],data7['Happiness Score'])
plt.title('Happiness Score by Year')
plt.show() 
#shows steady upward trend (other than 2017) for each year 
```

```python
low = min(data4['Economy (GDP per Capita)'])-.01
high = max(data4['Economy (GDP per Capita)'])+.01
plt.ylim(low, high)
plt.bar(data7["Years"],data4['Economy (GDP per Capita)'])
plt.title('Economy (GDP per Capita) Score by Year')
plt.show() 
#skyrockets in 2022 but other than that no real trend 
```

```python
low = min(data4['Family'])-.01
high = max(data4['Family'])+.01
plt.ylim(low, high)
plt.bar(data7["Years"],data4['Family'])
plt.title('Family Score by Year')
plt.show() 
#very low in 2016 and 2021 (very high 2017 to 2020)
```

```python
low = min(data4['Freedom'])-.01
high = max(data4['Freedom'])+.01
plt.ylim(low, high)
plt.bar(data7["Years"],data4['Freedom'])
plt.title('Freedom Score by Year')
plt.show() 
#shows upward trend, but 2016 and 2019 are outliers 
```

```python
low = min(data4['Health (Life Expectancy)'])-.01
high = max(data4['Health (Life Expectancy)'])+.01
plt.ylim(low, high)
plt.bar(data7["Years"],data4['Health (Life Expectancy)'])
plt.title('Health (Life Expectancy) Score by Year')
plt.show() 
#no real trend, but 2019 and 2020 were high 
```

```python
low = min(data4['Trust (Government Corruption)'])-.01
high = max(data4['Trust (Government Corruption)'])+.01
plt.ylim(low, high)
plt.bar(data7["Years"],data4['Trust (Government Corruption)'])
plt.title('Trust (Government Corruption) Score by Year')
plt.show() 
#parabola shaped where lowest in 2018 and 2019 and highest 2015 and 2022
```

```python
low = min(data4['Generosity'])-.01
high = max(data4['Generosity'])+.01
plt.ylim(low, high)
plt.bar(data7["Years"],data4['Generosity'])
plt.title('Generosity Score by Year')
plt.show() 
#shop decrease after 2017 and very low 2022
```
